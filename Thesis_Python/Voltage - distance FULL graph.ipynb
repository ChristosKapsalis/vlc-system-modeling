```python
import pandas as pd
import matplotlib.pyplot as plt
from tkinter import Tk, filedialog

Tk().withdraw()

file_path = filedialog.askopenfilename(
    title="Επέλεξε Excel αρχείο",
    filetypes=[("Excel files", "*.xlsx *.xls")]
)

if file_path:
    df = pd.read_excel(file_path)

    plt.figure(figsize=(8, 5))

    if {"D1", "V1", "D2", "V2", "D3", "V3"}.issubset(df.columns):

        plt.plot(df["D1"], df["V1"], marker="o", linestyle="-", label="y = 0 cm")

        plt.plot(df["D2"], df["V2"], marker="s", linestyle="-", label="y = 2 cm")

        plt.plot(df["D3"], df["V3"], marker="^", linestyle="-", label="y = 5 cm")

        plt.xlabel("Distance (cm)")
        plt.ylabel("Voltage (V)")
        plt.title("Voltage vs Distance για διαφορετικά y")
        plt.grid(True)
        plt.legend()
        plt.show()
    else:
        print("Το αρχείο δεν περιέχει τις στήλες D1,V1,D2,V2,D3,V3.")
else:
    print("Δεν επιλέχθηκε αρχείο.")
```
