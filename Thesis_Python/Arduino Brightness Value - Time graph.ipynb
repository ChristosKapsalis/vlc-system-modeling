```python
import pandas as pd
import matplotlib.pyplot as plt
from tkinter import Tk, filedialog

def plot_light_data_col3():
    """
#Ανοίγει αρχείο (CSV ή Excel), διαβάζει την 3η στήλη των επεξεργασμένων μετρήσεων φωτεινότητας)και φτιάχνει γράφημα με άξονα χρόνου (2 sec ανά μέτρηση)."""
    Tk().withdraw()
    file_path = filedialog.askopenfilename(
        title="Επιλέξτε αρχείο με μετρήσεις",
        filetypes=[("CSV αρχεία", "*.csv"), ("Excel αρχεία", "*.xlsx *.xls")]
    )

    if not file_path:
        print("Δεν επιλέχθηκε αρχείο.")
        return

    if file_path.lower().endswith(".csv"):
        df = pd.read_csv(file_path, header=None, usecols=[2])
    else:
        df = pd.read_excel(file_path, header=None, usecols=[2])

    values = pd.to_numeric(df.iloc[:,0], errors='coerce')
    values = values[(values >= 0) & (values <= 499)].dropna().astype(int).reset_index(drop=True)

    time = values.index * 2

    plt.figure(figsize=(10,5))
    plt.plot(time, values, marker='o', linestyle='-', markersize=3)
    plt.title("Μετρήσεις Φωτεινότητας (0-1023)")
    plt.xlabel("Χρόνος (sec)")
    plt.ylabel("Τιμή Φωτεινότητας")
    plt.grid(True)
    plt.show()

plot_light_data_col3()
```
