```python
import pandas as pd
import matplotlib.pyplot as plt
from tkinter import Tk, filedialog

Tk().withdraw()

file_path = filedialog.askopenfilename(
    title="Επέλεξε Excel αρχείο",
    filetypes=[("Excel files", "*.xlsx *.xls")]
)

if file_path:
    df = pd.read_excel(file_path)

    if "V" in df.columns and "D" in df.columns:
        plt.figure(figsize=(8, 5))
        plt.plot(df["D"], df["V"], marker="o", linestyle="-", color="b")

        plt.xlabel("Distance (D)")
        plt.ylabel("Voltage (V)")
        plt.title("Voltage vs Distance")
        plt.grid(True)
        plt.show()
    else:
        print("Το αρχείο δεν περιέχει τις στήλες 'V' και 'D'.")
else:
    print("Δεν επιλέχθηκε αρχείο.") 
```
