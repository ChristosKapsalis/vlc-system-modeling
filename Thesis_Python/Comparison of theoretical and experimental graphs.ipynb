```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
from tkinter import Tk, filedialog

def fit_and_plot_from_excel():
   
    Tk().withdraw()  # κρύβει το κεντρικό παράθυρο
    file_path = filedialog.askopenfilename(
        title="Επέλεξε το Excel αρχείο",
        filetypes=[("Excel Files", "*.xlsx *.xls")]
    )

    if not file_path:
        print("Δεν επιλέχθηκε αρχείο.")
        return

    df = pd.read_excel(file_path)

    labels = ["y=0cm", "y=2cm", "y=5cm"]

    plt.figure(figsize=(8,6))

    for i in range(1, 4):  
        D = df[f"D_{i}"].dropna().to_numpy()
        S = df[f"S_{i}"].dropna().to_numpy()
        V = df[f"V_{i}"].dropna().to_numpy()

        # V = aS + b
        res = linregress(S, V)
        a, b, R2 = res.slope, res.intercept, res.rvalue**2

        S_fit = np.linspace(min(S), max(S), 100)
        V_fit = a*S_fit + b

        plt.scatter(D, V, label=f"Πειρ. {labels[i-1]}", s=50)

        plt.plot(np.interp(S_fit, S[::-1], D[::-1]), V_fit,
                 label=f"Fit {labels[i-1]} (R²={R2:.3f})")

        print(f"{labels[i-1]}: a={a:.3f}, b={b:.3f}, R²={R2:.3f}")

    plt.xlabel("Απόσταση D [m]")
    plt.ylabel("Τάση V [Volt]")
    plt.title("Σύγκριση πειραματικών & θεωρητικών (Linear Fit από Excel)")
    plt.legend()
    plt.grid(True)
    plt.show()

fit_and_plot_from_excel()
```
