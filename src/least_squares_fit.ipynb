```python
import matplotlib.pyplot as plt

x = [0.25,0.26,0.27,0.28,0.29,0.30,0.31,0.32,0.33,0.34,0.35]

V_exp = [2.355,1.963,1.657,1.422,1.236,1.085,0.958,0.772,0.691,0.630,0.630]

S_theo = [16.000,14.792,13.717,12.755,11.872,11.111,10.406,9.766,9.182,8.650,8.163]

# without offset (V = k*S)
k = 0.1112
V_fit_no_offset = [s*k for s in S_theo]

# offset (V = alpha*S + beta)
alpha = 0.221
beta = -1.321
V_fit_linear = [alpha*s + beta for s in S_theo]

plt.figure(figsize=(8,5))
plt.plot(x, V_fit_no_offset, label='Fit χωρίς offset (k*S)', color='blue', linewidth=2)
plt.plot(x, V_fit_linear, label='Fit με offset (αS+β)', color='green', linewidth=2)
plt.scatter(x, V_exp, label='Πειραματικά V', color='red', s=50)
plt.xlabel('Απόσταση x [m]')
plt.ylabel('Τάση V [Volt]')
plt.title('Σύγκριση πειραματικών και θεωρητικών τιμών')
plt.legend()
plt.grid(True)
plt.show()
```
